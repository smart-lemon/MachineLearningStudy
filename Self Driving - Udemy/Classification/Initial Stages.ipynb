```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Creating the data set with the top and bottom array

```python
n_pts = 100
np.random.seed(0)
# The mean and standard deviation (of 2)
top_region    = np.array([np.random.normal(10, 2, n_pts), np.random.normal(12, 2, n_pts)]).T
bottom_region = np.array([np.random.normal(5, 2, n_pts), np.random.normal(6, 2, n_pts)]).T
```

Plotting of the data

```python
_, ax = plt.subplots(figsize = (4, 4))
ax.scatter(top_region[:,0], top_region[:,1], color = 'red')
ax.scatter(bottom_region[:, 0], bottom_region[:, 1], color = 'blue')

plt.show()
```
